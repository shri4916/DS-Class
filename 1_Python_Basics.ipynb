# Slicing of string
slice means a piece

operator is called slice operator, which can be used to retrieve parts of String.

In Python Strings follows zero based index.

The index can be either +ve or -ve.

+ve index means forward direction from Left to Right

-ve index means backward direction from Right to Left

```python
a  = 'Shriniwas'
# fetch sixth index value from start - 0,1,2,3,4,5
print (a[5])
# fetch third last index value - _______-3,-2,-1
print (a[-3])
# fetch first index value
print (a[0])
# fetch last index value
print (a[-1])
```

```python
# slice str using range
# fetch 4&5 index value
a  = 'Shriniwas'
print(a[3:5])
# extract first 4 letters
print(a[:4])
# fetch all letters
print(a[:])
# find length of string
len(a)
```
