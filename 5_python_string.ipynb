#Q)Write a Program to access each Character of String in Forward and Backward Direction by using while Loop?

```python
s = input("Enter a string :")
l = len(s)
i = 0
print("Foreward")
while i<l:
  print(s[i],end='')
  i+=1

i=l-1
print("\nBackward")
while i>=0:
  print(s[i],end='')
  i-=1
```

#Removing Spaces from the String:
1)rstrip()-->To remove spaces at right hand side.

lstrip()--> To remove spaces at left hand side.

strip()-->To remove spaces both sides.

```python
s = input("Enter city name :")
city = s.strip()
if city=="Pune":
  print("Welcome to Pune")
else:
  print("City not found")
```

```python
s = " data l"
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

#Finding Substrings:
We can use the following 4 methods

For forward direction:

1. find()

2. index()

For backward direction:

1. rfind()

2. rindex()

# find()

```python
# find(): Returns index of first occurrence of the given substring. If it is not available then we will
# get -1.
s="data"
s.find("data") #0th index
```

```python
#synntax - variable.find("Substring", start index, end index)
s="data"
print(s.find('a',1,2))
```

#2)index()

index() method is exactly same as find() method except that if the specified substring is not available then we will get ValueError.

```python
# we will get a error if the substring is not available
s.index('m')
```
