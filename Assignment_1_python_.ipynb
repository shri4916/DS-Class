# What is a List?

List is set of data elements in indexing format.

List can contain Int, float, boolean, complex, string

List is mutable means it can be edited

List is growable in nature

List allows Duplicates

List is initiated using []

```python
l = ["Shri", "Niwas", "Ambad"]
l[2] = "Rishi" # upadate value
l.append("Kale") #adding new value
print(l)
```

#4. Python Program to find the largest element in the list

```python
#Type 1 - max()

Numbers = [ 1, 32, 56, 87, 63]

largest = max(Numbers)
print(largest)
```

```python
#Type 2  - For loop

Numbers = [ 1, 32, 56, 87, 63]

largest = Numbers[0]

for num in Numbers:
  if num > largest:
    largest = num

print(largest)
```

```python
# Type 3 - Sort

Numbers = [ 1, 32, 56, 87, 63]

Numbers.sort()

print(Numbers[-1])
```

#5. Python program to interchange first and last elements in a list.

```python
#type 1 - simple assigment

l = [1,2,3,4,5]
lfirst = l[0]
l[0] = l[-1]
l[-1] = lfirst
print(l)
```

```python
#Type 2 - Swapping
l = [1,2,3,4,5]
l[0],l[-1] = l[-1],l[0]
print(l)
```

```python
#Type 3 - Using function


def swap_first_Last(l):
  l[0],l[-1] = l[-1],l[0]
  return(l)

l = [1,2,3,4,5]
print("List after swapping", swap_first_Last(l))
```

#6. Python program to swap two elements in a list

```python
#type 1 - swapping

l = [1,2,3,4,5]
pos1 = int(input("Enter first position : "))
pos2 = int(input("Enter second position : "))

l[pos1],l[pos2] = l[pos2],l[pos1]

print(l)
```

```python
# Type 2 - Using temp variable

l = [1,2,3,4,5]
pos1 = int(input("Enter first position : "))
pos2 = int(input("Enter second position : "))

temp = l[pos1]
l[pos1] = l[pos2]
l[pos2] = temp

print(l)
```

```python
#Type 3 - Using function
l = [1,2,3,4,5]
pos1 = int(input("Enter first position : "))
pos2 = int(input("Enter second position : "))

def swap_position(l, pos1, pos2):
  l[pos1],l[pos2] = l[pos2],l[pos1]
  return(l)

print(swap_position(l,pos1,pos2))
```
