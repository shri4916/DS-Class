#2) Iterative Statements:
 If we want to execute a group of statements multiple times then we should go for Iterative statements.

Python supports 2 types of iterative statements.

1. for loop

2. while loop

 for loop:

If we want to execute some action for every element present in some sequence (it may be string or collection) then we should use for loop.

Syntax:

for x in sequence:

Body

Where sequence can be string or any collection.

Body will be executed for every element present in the sequence.

```python
# To print characters present in the given string
a = "Shri"
for x in a:
  print(x,end='')
```

```python
# To print Hello 10 times
for x in range(10):
  print("Hare Krishna")
```

```python
# To display odd numbers from 0 to 20
for x in range(20):
  if x%2 != 0:
    print(x)
```

```python
# To display numbers from 10 to 1 in descending order
for x in range(10,0,-1):
  print(x)
```

```python
# To print sum of numbers presenst inside list given by cmd
a = eval(input("Enter numbers :"))
lsum = 0
for x in a:
  lsum = lsum+int(x)

print(lsum)
```

#2) While Loop
If we want to execute a group of statements iteratively until some condition become false,then we should go for while loop.

```python
# To print numbers from 1 to 10 by using while loop
x = 1
while x <= 10:
  print (x)
  x+=1
```

```python
# To display the sum of first n numbers
n = int(input("Enter Nth number :"))
x = 1
add = 0
while x <=n:
 add += x
 x+=1

print("Sum of first ",n, "th number is :",add)
```

# Nested Loop

```python

# we can take a loop inside another loop,which are also known as nested loops.
for i in range(4):
  for j in range(4):
    print("I =",i,"J =",j)
```
