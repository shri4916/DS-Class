#TYPE CASTING
We can convert one type value to another type. This conversion is called Typecasting or Type coersion.

֍ The following are various inbuilt functions for type casting.

int()

float()

complex()

bool()

str()

# int():

We can use this function to convert values from other types to int
We can convert
1. float
2. boolean
3. string with int numbers (Not float)

```python
int(1.09)
```

```python
print(int(10))
print(int("10"))
print(int(True))
print(int(False))
print(int("10.08"))
```

#float():
We can use float() function to convert other type values to float type. (We can not convert string value containing letters into float but if string value is number, we can convert it into a string)
 we can convert following types into float
 1. int
 2. boolean
 3. string (Only with numbers)

```python
print(float(10))
print(float("10"))
print(float(True))
print(float(False))
print(float("10.08"))
```

#complex():
We can use complex() function to convert other types to complex type.

Form-1: complex(x)

We can use this function to convert x into complex number with real part x and imaginary part 0.

```python
print(complex(12))
print(complex(True))
print(complex(False))
```

#Form-2: complex(x,y)

We can use this method to convert x and y into complex number such that x will be real part and y will be imaginary part.

Eg: complex(10, -2)  10-2j

complex(True, False)  1+0j



```python
print(complex(10,2))
print(complex(True,False))
```

#bool():

#8) List Data Type:
If we want to represent a group of values as a single entity where insertion order required to preserve and duplicates are allowed then we should go for list data type.

Insertion Order is preserved

Heterogeneous Objects are allowed

Duplicates are allowed

Growable in nature

Values should be enclosed within square brackets.

```python
a = [1,2,2,3,4,"S",10.2,True,"sd",1+0j]
print(a)
#add new values to list
a.append(10.7)
print(a)
#update values
a[0]="Start"
print(a)
#delete or remove values
a.remove(10.2)
print(a)
#we can duble the list by multipling
a = a*2
print(a)
```

#10)Range Data Type:
 range Data Type represents a sequence of numbers.

 The elements present in range Data type are not modifiable. i.e range Data type is immutable.

```python
#Syntax - range(starting , ending , step size)
a = range(10,20,2)
print(a)
```

```python
for i in a:
  print(i)
```

```python
# we can fetch data in range using index
print(a[0] )

# we ccan not assing new values in range
a[20] = 100
```

```python
#we can create a list from range
s = range(20)
print(type(s))
a = list(s)
print(a)
print(type(a))
```

#11) set Data Type:
֍ If we want to represent a group of values without duplicates where order is not important then we should go for set Data Type.

Insertion order is not preserved

Duplicates are not allowed

Heterogeneous objects are allowed

Index concept is not applicable

It is mutable collection

Growable in nature

```python
s={100,0,10,200,10,'Data Science'}
```

```python
# set does not support indexing
s[0]
```

```python
#Set is growable in nature, we can add remove itemms
s.add(700)
print(s)
s.remove(10)
print(s)
```

#13) dict Data Type:
